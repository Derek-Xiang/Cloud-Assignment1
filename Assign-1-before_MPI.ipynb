```python
block_bounds=[]
block_counts= Counter()
hashtag_counts= {}

# start = time.time()
# end = time.time()
# print('time consume: '+str(end-start))
```

```python
for feature in grid["features"]:
    # get all the bounds of the 16 zones
#     block_bounds[feature['properties']['id']] = feature['properties']
    block_bounds.append(feature['properties'])
    # initialize the counters of blocks
    block_counts[feature['properties']['id']] = 0
    hashtag_counts[feature['properties']['id']] = Counter()
```

```python
'''
returns which block this coordinate sits in
Input: coordinate --> [x,y]
Ourput: block --> str
'''
def get_block(pos):
    for bound in block_bounds:
        if bound['xmin'] < pos[0] <= bound['xmax']\
            and bound['ymin'] < pos[1] <= bound['ymax']:
            return bound['id']
        else:
            # if not sit within the first block, check next
            continue
```
